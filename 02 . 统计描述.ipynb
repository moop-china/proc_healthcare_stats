```python
df = pd.read_csv(dir+"data/心脏病诊断数据集（原数据）.csv",header=None) #读取csv文件，没有列名
df.columns = ["年龄","性别","胸痛类型","静息血压","血浆类固醇含量（mg/dl）","空腹血糖>120mg/dl","静息心电图结果","最高心率","运动型心绞痛","运动引起的ST下降","最大运动量时心电图ST的斜率","使用荧光染色法测定的主血管数","THAL","患病情况"]                                                 #添加列名
df.head(5)
```
